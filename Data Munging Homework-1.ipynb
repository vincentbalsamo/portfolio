### Problem 4 Data pre-processing

a) Load mtcars.csv into notebook as a new dataframe "cars". Change the name of the first column as "car_names".

```python
#write your code here
cars=pd.read_csv('mtcars.csv',sep=',')
cars.columns=["car_names","mpg","cyl","disp","hp","drat","wt","qsec","vs","am","gear","carb"]
cars
```

b) Using lambda function to create a new column 'output1' to return the reciprocal of horsepower. Create another column 'output2' to return mpg/hp.

```python
#write your code here
cars['output1']=cars.apply(lambda x:1/x['hp'],axis=1)
cars['output2']=cars.apply(lambda x:x['mpg']/x['hp'],axis=1)
cars
```

c) Group the data based on its cylinder and aggregate based on the minimum and maximun hourspower. 

```python
#write your code here
funcs = {'hp':['min','max']}
min_max_hp = cars.groupby(['cyl']).agg(funcs)
min_max_hp
```
